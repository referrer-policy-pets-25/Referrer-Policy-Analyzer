# Intersection

Find initial url from three vantage points. 

```python
import pandas as pd
```

```python
# change the path to the location of the parquet file
ams = pd.read_parquet('../output/data_raw/2025-02-08_noAct_raw.parquet')
sf = pd.read_parquet('../output/data_raw/2025-02-08_noAct_raw.parquet')
sg = pd.read_parquet('../output/data_raw/2025-02-08_noAct_raw.parquet')
```

```python
init1 = set(ams['init_url'])
init2 = set(sf['init_url'])
init3 = set(sg['init_url'])
```

```python
intersection = init1.intersection(init2).intersection(init3)
```

```python
intersection
```

```python
file_path = "../helpers/intersection"  

with open(file_path, "w") as file:
    for url in intersection:
        file.write(url + "\n")
```
